```python
# %matplotlib notebook
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
used_string = 'project-1/data_files/final.csv'
movie_files = pd.read_csv(used_string)

movie_files
```

```python
# Create a group based on the values in the 'rated' column
new_df=movie_files.loc[:,["rated","Infl. Adj. Dom. Box Office"]]
maker_rated = new_df.groupby('rated').mean()

# mean = maker_rated['Infl. Adj. Dom. Box Office'].mean()

# # Count how many times each rating appears in our group
# # count_rated = maker_group['rated'].count()

# # count_rated
# mean
maker_rated
```
